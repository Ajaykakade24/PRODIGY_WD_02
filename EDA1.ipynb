```python
import pandas as pd

ipl_df = pd.read_csv('/content/class_ipl_22-26_dataset.csv')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#1.which ipl teams scored the highest runs
```

```python
team_avg_runs = (
    ipl_df.groupby('team1')['team1_runs']
    .mean()
    .sort_values(ascending=False)
)

print("Average Runs by Team:")
print(team_avg_runs)

print("\nTeam with Highest Average Runs:")
print(team_avg_runs.head(1))
```

```python
plt.figure(figsize=(10,5))

team_avg_runs.head(10).plot(
    kind='bar'
)

plt.title('Top Teams by Average Team1 Runs')
plt.xlabel('Teams')
plt.ylabel('Average Runs')

plt.show()
```

```python
print("Mean :", ipl_df['team1_runs'].mean())

print("Median :", ipl_df['team1_runs'].median())

print("Mode :", ipl_df['team1_runs'].mode()[0])
```

```python
print("Minimum :", ipl_df['team1_runs'].min())

print("Maximum :", ipl_df['team1_runs'].max())

print("Standard Deviation :",
      ipl_df['team1_runs'].std())

print("Variance :",
      ipl_df['team1_runs'].var())
```

```python
#q6.misssing value analysis
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,6))

sns.heatmap(
    ipl_df.isnull(),
    cbar=True,
    yticklabels=False,
    cmap='viridis'
)

plt.title("Missing Values Heatmap")

plt.show()
```

```python
#Q7.Analyze missing values in the ipl dataset
```

```python
missing_count =ipl_df.isnull().sum()

missing_percentage = (
    ipl_df.isnull().sum() /
    len(ipl_df)
) * 100

print("Missing Count")
print(missing_count)

print("\nMissing Percentage")
print(missing_percentage)

sns.heatmap(ipl_df.isnull(),
            cbar=False)

plt.show()
```

```python
#8.which stadiums produce high scoring matches
```

```python
avg_runs =ipl_df.groupby(
    'venue'
)['team1_runs'].mean()

print(avg_runs.sort_values(
    ascending=False
))

plt.figure(figsize=(10,5))

sns.barplot(
    x=avg_runs.index,
    y=avg_runs.values
)

plt.xticks(rotation=90)
plt.title("Average Team1 Runs by Venue")
plt.show()

sns.boxplot(
    x='venue',
    y='team1_runs',
    data=ipl_df
)

plt.xticks(rotation=90)
plt.show()
```

```python
#9.multivariate Analsis of match outcomes
```

```python
sns.scatterplot(
    x='team1_runs',
    y='team2_runs',
    hue='winner',
    data=ipl_df
)

plt.title("Runs vs Winner")
plt.show()

sns.pairplot(
    ipl_df,
    vars=[
        'team1_runs',
        'team2_runs',
        'team1_wickets',
        'team2_wickets'
    ],
    hue='winner'
)

plt.show()
```
