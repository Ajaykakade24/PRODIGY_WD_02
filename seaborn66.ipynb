```python
ipl_df = pd.read_csv('/content/class_ipl_22-26_dataset.csv')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))

sns.countplot(
    y='winner',
    data=ipl_df,
    order=ipl_df['winner'].value_counts().index[:5]
)

plt.title("Top 5 Winning Teams")
plt.xlabel("Wins")
plt.ylabel("Teams")

plt.show()
```

```python
#q2
```

```python
season_avg = ipl_df.groupby('season')['team1_runs'].mean().reset_index()

plt.figure(figsize=(12,6))

sns.barplot(
    x='season',
    y='team1_runs',
    data=season_avg
)

plt.title("Season Wise Average Team1 Score")
plt.xlabel("Season")
plt.ylabel("Average Runs")

plt.xticks(rotation=45)

plt.show()
```

```python
#q3
```

```python
plt.figure(figsize=(10,6))

sns.scatterplot(
    x='team1_runs',
    y='team2_runs',
    hue='winner',
    data=ipl_df
)

plt.title("Team1 Runs vs Team2 Runs")

plt.show()
```

```python
plt.figure(figsize=(15,7))

sns.boxplot(
    x='venue',
    y='team1_runs',
    data=ipl_df
)

plt.xticks(rotation=90)

plt.title("Venue vs Team1 Runs")

plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.kdeplot(
    ipl_df['team1_runs'],
    label='Team1 Runs',
    fill=True
)

sns.kdeplot(
    ipl_df['team2_runs'],
    label='Team2 Runs',
    fill=True
)

plt.legend()

plt.title("Team1 vs Team2 Runs Distribution")

plt.show()
```

```python
numeric_cols = [
    'team1_runs',
    'team2_runs',
    'team1_wickets',
    'team2_wickets'
]

corr = ipl_df[numeric_cols].corr()

plt.figure(figsize=(8,6))

sns.heatmap(
    corr,
    annot=True,
    cmap='coolwarm'
)

plt.title("Correlation Heatmap")

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(18,12))

# 1. Count Plot - Winning Teams
plt.subplot(2,3,1)
sns.countplot(
    y='winner',
    data=ipl_df,
    order=ipl_df['winner'].value_counts().index[:5]
)
plt.title("Top 5 Winning Teams")

# 2. Bar Plot - Season Average Score
plt.subplot(2,3,2)
season_avg = ipl_df.groupby('season')['team1_runs'].mean().reset_index()
sns.barplot(
    x='season',
    y='team1_runs',
    data=season_avg
)
plt.xticks(rotation=45)
plt.title("Season Wise Avg Team1 Runs")

# 3. Histogram
plt.subplot(2,3,3)
sns.histplot(
    ipl_df['team1_runs'],
    kde=True,
    bins=20
)
plt.title("Team1 Runs Distribution")

# 4. Scatter Plot
plt.subplot(2,3,4)
sns.scatterplot(
    x='team1_runs',
    y='team2_runs',
    data=ipl_df
)
plt.title("Team1 vs Team2 Runs")

# 5. Heatmap
plt.subplot(2,3,5)

corr_cols = [
    'team1_runs',
    'team2_runs',
    'team1_wickets',
    'team2_wickets'
]

sns.heatmap(
    ipl_df[corr_cols].corr(),
    annot=True,
    cmap='coolwarm'
)
plt.title("Correlation Heatmap")

plt.tight_layout()
plt.show()
```

```python
IPL ANALYTICS REPORT

Objective:
The objective of this analysis is to study IPL match data using various visualization techniques and identify trends related to team performance, scoring patterns, venue impact, and match outcomes.

1. Most Successful Teams:
   The count plot of winning teams shows the top-performing franchises in IPL history. These teams have consistently achieved victories across multiple seasons and demonstrated strong overall performance.

2. Season-wise Average Scores:
   The bar plot of average Team 1 runs by season indicates how batting trends have evolved over the years. Some seasons recorded significantly higher average scores, suggesting batting-friendly conditions and aggressive gameplay.

3. Run Distribution Analysis:
   The histogram of Team 1 runs shows that most innings scores are concentrated within a specific range. This range represents the most common competitive score in IPL matches.

4. Team1 vs Team2 Score Relationship:
   The scatter plot reveals the relationship between first-innings and second-innings scores. High-scoring first innings often lead to high-scoring chases, although match situations and pitch conditions influence outcomes.

5. Correlation Analysis:
   The heatmap highlights relationships among runs and wickets. Strong positive correlations indicate variables that tend to increase together, while negative correlations suggest inverse relationships.

Conclusion:
IPL data analysis reveals that team performance, batting strength, venue conditions, and match strategies significantly affect outcomes. Visualization techniques such as count plots, bar plots, scatter plots, histograms, and heatmaps provide valuable insights into team success, scoring behavior, and overall tournament trends.
```
